# Data Visualization

```python
df['Exited'] = df['Exited'].map({0: 'Not_Churned', 1: 'Churned'})
sns.countplot(data=df, x='Gender', hue='Exited')
plt.show()
```

- Inference

Above Plot represents the no of male and female customers that can be churn or not churned.

It shows that count of male customers is higher than female customer for loan approval and count of male customers is less than female customers for not getting loan.

```python
sns.boxplot(data=df, x='Geography', y='CreditScore', hue='Exited')
plt.show()
```

- Inference

Box plot represents the creditscore of all customers according to geography that are churned or not as we can see that customer form all the regions having valid score for getting churned but we can also infer that there are some outliers that some customers having very less credit score which is below 400.

```python
sns.barplot(data=df, x='Geography', y='Balance')
plt.show()
```

```python
sns.barplot(data=df, x='Gender', y='EstimatedSalary', hue='Exited')
plt.show()
```

```python
sns.pairplot(df, vars=['CreditScore', 'Age', 'Balance', 'EstimatedSalary'], hue='Exited')
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Exited', kde=True)
plt.show()
```

- Inference 

The plot represents the count of churning based on the age of customers we can clearly infer that rate of churning follows normal distribution from the age group of 20 to 70.
