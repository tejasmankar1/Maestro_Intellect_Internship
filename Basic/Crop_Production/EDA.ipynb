# Bivariate analysis


```python
plt.figure(figsize=(25,15))
sns.barplot(x=df["State_Name"],y=df["Production"])
plt.xticks(rotation=90)
```

```python
sns.lineplot(x=df['Crop_Year'],y=df['Production'])
```

```python
plt.figure(figsize=(25,15))
sns.barplot(x=df["Crop_Year"],y=df["Production"])
plt.xticks(rotation=90)
```

- Inference: 

Line Plot and Bar Plot describes Production rate of Crops of different years.
It was very low during the year 1997 but afterwards it increased to a certain
production level and continued. In the year 2011 there was drastic increase in the Crop
Production and it continued till 2014. After that in the year 2015 there was a huge amount
of downfall in the Crop Production. Previously we have observed downfall on variety of Crops 
in the same year. This concludes that maybe because of some unexpected incidents or policies,
many types of Crops were not Produced which caused a drastic downfall in the Production rate.

```python
sns.barplot(x=df['Season'],y=df['Production'])
```

- Inference

The figure represents the production rate  with respect to season.

The production rate is higher in whole year than other respected sesons.


```python
plt.figure(figsize=(35,30))
sns.barplot(x=df["Crop"],y=df["Production"])
plt.xticks(rotation=90)
```

- Inference: 

The plot represents the production of coconuts is the the highest value when compared to other crops.
we can also see a small value of production of sugarcanes.

```python
plt.figure(figsize=(25,15))
sns.barplot(x=df["State_Name"],y=df["Area"])
plt.xticks(rotation=90)
```

- Inference 

From the above visualization we can observe the state wise area of agriculture land. 
We can infer that Punjab has the largest area for agriculture and maharashtra 
has the second largest and west bengal has the third largest area for agriculture.
We can also infer that chandigarh has the smallest area for agriculture.

```python
sns.lineplot(x=df["Crop_Year"],y=df["Area"])
```

- Inference 

The above visualization shows us the agriculture area in different
year intervals. We can infer from the visualization that the agriculture 
area was the largest during the year 1997 right after that we can see a 
drastic decline in the agriculture area from the year 2000. We can see 
further declining of the area from the year 2015

```python
sns.barplot(x=df["Season"],y=df["Area"])
```

- Inference 

The above visualization shows us the agriculture area in different 
seasons. We can infer from the visualization that in the winter season the 
Agriculture area is the largest with value of 35000. We can also infer that 
during summer season the agriculture area is very less.

```python
plt.figure(figsize=(25,15))
sns.barplot(x=df["Crop"],y=df["Area"])
plt.xticks(rotation=90)
```

- Inference 

The above visualisation tells us about the agriculture area
    alotted to different crops. From the visualization we can infer that 
    oilseeds have the maximum area and the black line above the barplot
    indicates that it might rise even more high. We can also infer that
    few crops have not been alotted any agriculture area as well.

```python
plt.figure(figsize=(25,15))
sns.barplot(x=df["Crop_Year"],y=df["Area"])
plt.xticks(rotation=90)
```

- Inference

 The above visualization shows us the Agriculture area for different
    crop years. From this we can infer that the year 1997 has the largest 
    agriculture area and it has been declining since then. In the visualization
