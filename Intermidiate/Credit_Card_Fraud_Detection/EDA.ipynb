# Data visualization

```python
sns.countplot(x='Class', data=df, palette='pastel')
plt.title('Fraud vs Non-Fraud Transactions')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

- Inference


The count plot provides the information of non fraud transactions is much higher than Fraud transactions.

This plot shows the imbalance in the dataset, where the majority of transactions are non-fraud.

```python
sns.histplot(df['Amount'], bins=50,  kde=True, color='blue')
plt.title('Transaction Amount Distribution')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()
```

- Inference

The distribution shows the presence of high value outliers for most transactions which are of low amounts.

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False, cbar=True)
plt.title('Feature Correlation Heatmap')
plt.show()

```

- Inference

The correlation matrix shows features with strong correlations (positive or negative) to Class might be significant predictors of fraud.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Class', y='Amount', data=df, palette='Set2')
plt.title('Transaction Amounts by Class')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Transaction Amount')
plt.show()
```

- Inference

Presence of outliers in Fraud tranction might indicate deliberate attempts to bypass detection.

```python
sns.boxplot(x='Class', y='Time', data=df, palette='coolwarm')
plt.title('Transaction Time Distribution by Class')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Time (Seconds)')
plt.show()
```

- Inference

Represents whether fraudulent transactions occur predominantly at specific times compared to non-fraudulent ones.

```python
fraud_data = df[df['Class'] == 1]

plt.figure(figsize=(20, 10))
sns.heatmap(fraud_data.corr(), cmap='coolwarm', annot=False)
plt.title('Feature Correlation Heatmap (Fraud Only)')
plt.show()
```

- Inference

Correlation analysis specifically for fraud transactions.

```python
fraud_data = df[df['Class'] == 0]

plt.figure(figsize=(20, 10))
sns.heatmap(fraud_data.corr(), cmap='coolwarm', annot=False)
plt.title('Feature Correlation Heatmap (Non-Fraud Only)')
plt.show()
```

- Inference

Correlation analysis specifically for Non-fraud transactions.

```python
for column in df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[column], kde=True, bins=30)
    plt.title(f'Distribution of {column}')
    plt.show()
```
